```python
# ========== ЯЧЕЙКА 4 ==========
html_part4 = '''
<!-- СЛАЙД 11 -->
<div class="slide" id="s11">
<div class="slide-title">Ключевые выводы</div>
<div class="slide-sub">Результаты проекта «AI-Конвейер юридического контента: Трудовое право 2026»</div>
<div class="divider"></div>
<div style="display:grid;grid-template-columns:1fr 1fr;gap:14px;width:100%;">
  <div>
    <ul class="check-list">
      <li><span class="ico">✅</span><div><strong>Промпт-чейнинг сокращает время в 20 раз</strong><br><span style="font-size:11px;color:rgba(255,255,255,.65);">8 инструментов в связке дают результат за 2-3 часа вместо 40 часов ручного труда</span></div></li>
      <li><span class="ico">✅</span><div><strong>Юридическая точность требует проверки</strong><br><span style="font-size:11px;color:rgba(255,255,255,.65);">AI генерирует контент быстро, но финальную верификацию НПА должен делать юрист</span></div></li>
      <li><span class="ico">✅</span><div><strong>GigaChat лучше для русскоязычного права</strong><br><span style="font-size:11px;color:rgba(255,255,255,.65);">Понимает контекст российского законодательства лучше западных моделей</span></div></li>
      <li><span class="ico">✅</span><div><strong>Obsidian структурирует всю базу знаний</strong><br><span style="font-size:11px;color:rgba(255,255,255,.65);">Граф связей между заметками ускоряет навигацию и повторное использование промптов</span></div></li>
      <li><span class="ico">✅</span><div><strong>Python-визуализация усиливает контент</strong><br><span style="font-size:11px;color:rgba(255,255,255,.65);">Графики из реальных данных Роструда повышают доверие читателей и экспертность</span></div></li>
    </ul>
  </div>
  <div style="display:flex;flex-direction:column;gap:12px;">
    <div style="background:rgba(244,185,66,.1);border:1px solid var(--border);border-radius:12px;padding:16px;">
      <div style="font-size:13px;font-weight:700;color:var(--gold);margin-bottom:10px;">📈 Итоговые цифры проекта:</div>
      <div style="display:grid;grid-template-columns:1fr 1fr;gap:10px;">
        <div style="text-align:center;background:rgba(0,0,0,.2);border-radius:10px;padding:10px;">
          <div style="font-size:28px;font-weight:900;color:var(--gold);">8</div>
          <div style="font-size:10px;color:rgba(255,255,255,.6);">AI-инструментов</div>
        </div>
        <div style="text-align:center;background:rgba(0,0,0,.2);border-radius:10px;padding:10px;">
          <div style="font-size:28px;font-weight:900;color:var(--gold);">11</div>
          <div style="font-size:10px;color:rgba(255,255,255,.6);">промптов</div>
        </div>
        <div style="text-align:center;background:rgba(0,0,0,.2);border-radius:10px;padding:10px;">
          <div style="font-size:28px;font-weight:900;color:var(--gold);">×20</div>
          <div style="font-size:10px;color:rgba(255,255,255,.6);">ускорение работы</div>
        </div>
        <div style="text-align:center;background:rgba(0,0,0,.2);border-radius:10px;padding:10px;">
          <div style="font-size:28px;font-weight:900;color:var(--gold);">6</div>
          <div style="font-size:10px;color:rgba(255,255,255,.6);">форматов контента</div>
        </div>
      </div>
    </div>
    <div style="background:rgba(100,220,100,.08);border:1px solid rgba(100,220,100,.3);border-radius:12px;padding:14px;">
      <div style="font-size:13px;font-weight:700;color:#90ee90;margin-bottom:8px;">🎯 Навык будущего:</div>
      <div style="font-size:12px;color:rgba(255,255,255,.8);line-height:1.7;">
        Специалист по ИИ, умеющий выстраивать промпт-чейнинг — это новая профессия, позволяющая автоматизировать любую экспертную область: право, медицину, финансы
      </div>
    </div>
    <div style="background:rgba(244,185,66,.08);border:1px solid var(--border);border-radius:12px;padding:12px;text-align:center;">
      <div style="font-size:12px;color:var(--gold);font-weight:700;">🏛️ РАНХиГС | Май 2026</div>
      <div style="font-size:11px;color:rgba(255,255,255,.6);margin-top:4px;">Специалист по системам искусственного интеллекта</div>
    </div>
  </div>
</div>
</div>

<!-- СЛАЙД 12 -->
<div class="slide" id="s12">
<div class="slide-title">Контент-пакет готов к публикации!</div>
<div class="slide-sub">Проект завершён | Все материалы подготовлены и структурированы</div>
<div class="divider"></div>
<div style="display:grid;grid-template-columns:1.2fr 0.8fr;gap:16px;width:100%;">
  <div>
    <div style="font-size:13px;font-weight:700;color:var(--gold);margin-bottom:10px;">📦 Итоговый контент-пакет включает:</div>
    <div style="display:grid;grid-template-columns:1fr 1fr;gap:10px;margin-bottom:14px;">
      <div style="background:var(--card);border:1px solid var(--border);border-radius:10px;padding:12px;text-align:center;">
        <div style="font-size:26px;margin-bottom:6px;">📰</div>
        <div style="font-size:12px;font-weight:700;color:var(--gold);">Статья</div>
        <div style="font-size:10px;color:rgba(255,255,255,.6);">5200 слов | GigaChat</div>
      </div>
      <div style="background:var(--card);border:1px solid var(--border);border-radius:10px;padding:12px;text-align:center;">
        <div style="font-size:26px;margin-bottom:6px;">❓</div>
        <div style="font-size:12px;font-weight:700;color:var(--gold);">FAQ</div>
        <div style="font-size:10px;color:rgba(255,255,255,.6);">10 вопросов | ChatGPT</div>
      </div>
      <div style="background:var(--card);border:1px solid var(--border);border-radius:10px;padding:12px;text-align:center;">
        <div style="font-size:26px;margin-bottom:6px;">📱</div>
        <div style="font-size:12px;font-weight:700;color:var(--gold);">3 Поста</div>
        <div style="font-size:10px;color:rgba(255,255,255,.6);">VK + TG + Instagram</div>
      </div>
      <div style="background:var(--card);border:1px solid var(--border);border-radius:10px;padding:12px;text-align:center;">
        <div style="font-size:26px;margin-bottom:6px;">📊</div>
        <div style="font-size:12px;font-weight:700;color:var(--gold);">5 Графиков</div>
        <div style="font-size:10px;color:rgba(255,255,255,.6);">Python | Colab</div>
      </div>
      <div style="background:var(--card);border:1px solid var(--border);border-radius:10px;padding:12px;text-align:center;">
        <div style="font-size:26px;margin-bottom:6px;">🖼️</div>
        <div style="font-size:12px;font-weight:700;color:var(--gold);">Инфографика</div>
        <div style="font-size:10px;color:rgba(255,255,255,.6);">SVG | Nana Banana</div>
      </div>
      <div style="background:var(--card);border:1px solid var(--border);border-radius:10px;padding:12px;text-align:center;">
        <div style="font-size:26px;margin-bottom:6px;">🎨</div>
        <div style="font-size:12px;font-weight:700;color:var(--gold);">Презентация</div>
        <div style="font-size:10px;color:rgba(255,255,255,.6);">12 слайдов | HTML</div>
      </div>
    </div>
    <div style="background:rgba(244,185,66,.12);border:1px solid var(--gold);border-radius:12px;padding:12px;text-align:center;">
      <span style="font-size:13px;font-weight:700;color:var(--gold);">
        🏆 Всё создано за ~2.5 часа с помощью AI-конвейера
      </span>
    </div>
  </div>
  <div style="display:flex;flex-direction:column;gap:12px;align-items:center;">
    <div style="background:var(--card);border:2px solid var(--gold);border-radius:14px;padding:20px;text-align:center;width:100%;">
      <div style="font-size:13px;font-weight:700;color:var(--gold);margin-bottom:10px;">📂 Проект на GitHub:</div>
      <div style="background:rgba(0,0,0,.3);border-radius:10px;padding:12px;margin-bottom:10px;">
        <div style="font-size:28px;margin-bottom:6px;">🐙</div>
        <a href="https://github.com/inyanochka-bit/AI--.-."
           target="_blank"
           style="color:#f4b942;font-size:11px;font-weight:700;word-break:break-all;text-decoration:none;line-height:1.6;">
          github.com/inyanochka-bit/<br>AI--.-.
        </a>
      </div>
      <div style="background:rgba(244,185,66,.1);border-radius:8px;padding:8px;">
        <div style="font-size:10px;color:rgba(255,255,255,.6);">Все промпты, код и материалы</div>
        <div style="font-size:10px;color:rgba(255,255,255,.6);">открыты для скачивания</div>
      </div>
    </div>
    <div style="background:var(--card);border:1px solid var(--border);border-radius:12px;padding:14px;width:100%;text-align:center;">
      <div style="font-size:22px;margin-bottom:8px;">👩‍💻</div>
      <div style="font-size:13px;font-weight:700;color:var(--gold);">Ерастова Яна Викторовна</div>
      <div style="font-size:11px;color:rgba(255,255,255,.7);margin-top:4px;">РАНХиГС</div>
      <div style="font-size:11px;color:rgba(255,255,255,.7);">Специалист по системам ИИ</div>
      <div style="margin-top:10px;padding-top:10px;border-top:1px solid var(--border);">
        <div style="font-size:11px;color:rgba(255,255,255,.5);">Май 2026</div>
        <div style="font-size:11px;color:rgba(255,255,255,.5);">ИИ в профессиональной деятельности</div>
      </div>
    </div>
  </div>
</div>
</div>

</div><!-- конец #wrap -->

<script>
(function(){
  var total = 12;
  var cur = 0;
  var slides = document.querySelectorAll(".slide");
  var cntEl = document.getElementById("cnt");
  var progEl = document.getElementById("prog");
  var startX = 0;
  var isDragging = false;

  function goTo(n){
    if(n < 0 || n >= total) return;
    slides[cur].classList.remove("active");
    slides[cur].classList.add("exit");
    var prev = cur;
    cur = n;
    slides[cur].classList.add("active");
    setTimeout(function(){ slides[prev].classList.remove("exit"); }, 460);
    cntEl.textContent = (cur+1) + " / " + total;
    progEl.style.width = ((cur+1)/total*100) + "%";
  }

  document.getElementById("btn-n").addEventListener("click", function(){ goTo(cur+1); });
  document.getElementById("btn-p").addEventListener("click", function(){ goTo(cur-1); });

  document.addEventListener("keydown", function(e){
    if(e.key==="ArrowRight"||e.key==="ArrowDown") goTo(cur+1);
    if(e.key==="ArrowLeft" ||e.key==="ArrowUp")   goTo(cur-1);
    if(e.key==="F11"){ e.preventDefault(); toggleFS(); }
  });

  document.addEventListener("touchstart",function(e){ startX=e.touches[0].clientX; isDragging=true; },{passive:true});
  document.addEventListener("touchend",  function(e){
    if(!isDragging) return;
    var dx=e.changedTouches[0].clientX-startX;
    if(Math.abs(dx)>50){ dx<0?goTo(cur+1):goTo(cur-1); }
    isDragging=false;
  },{passive:true});

  document.addEventListener("mousedown",function(e){ startX=e.clientX; isDragging=true; });
  document.addEventListener("mouseup",  function(e){
    if(!isDragging) return;
    var dx=e.clientX-startX;
    if(Math.abs(dx)>60){ dx<0?goTo(cur+1):goTo(cur-1); }
    isDragging=false;
  });

  function toggleFS(){
    if(!document.fullscreenElement){ document.documentElement.requestFullscreen&&document.documentElement.requestFullscreen(); }
    else{ document.exitFullscreen&&document.exitFullscreen(); }
  }
  document.getElementById("fs-btn").addEventListener("click",toggleFS);
})();
</script>
</body>
</html>
'''

# =====================================================
# СБОРКА ПОЛНОГО HTML
# =====================================================
full_html = html_part1 + html_part2 + html_part3 + html_part4

# Сохраняем HTML файл
html_filename = "erastova_presentation_2026.html"
with open(html_filename, "w", encoding="utf-8") as f:
    f.write(full_html)

print(f"✅ HTML создан: {html_filename}")

# =====================================================
# УСТАНОВКА ЗАВИСИМОСТЕЙ ДЛЯ PDF
# =====================================================
print("📦 Устанавливаю зависимости для PDF...")
import subprocess
subprocess.run(["apt-get", "install", "-y", "-q", "chromium-browser"], capture_output=True)
subprocess.run(["pip", "install", "-q", "selenium", "webdriver-manager"], capture_output=True)
print("✅ Зависимости установлены")

# =====================================================
# КОНВЕРТАЦИЯ В PDF ЧЕРЕЗ PLAYWRIGHT (ЛУЧШИЙ СПОСОБ)
# =====================================================
print("📦 Устанавливаю Playwright...")
subprocess.run(["pip", "install", "-q", "playwright"], capture_output=True)
subprocess.run(["python", "-m", "playwright", "install", "chromium"], capture_output=True)
subprocess.run(["python", "-m", "playwright", "install-deps", "chromium"], capture_output=True)
print("✅ Playwright готов")

import asyncio
import os
from pathlib import Path

async def make_pdf():
    from playwright.async_api import async_playwright

    pdf_filename = "erastova_presentation_2026.pdf"
    html_path = Path(html_filename).resolve()

    async with async_playwright() as p:
        browser = await p.chromium.launch(
            headless=True,
            args=["--no-sandbox","--disable-dev-shm-usage","--disable-gpu"]
        )
        page = await browser.new_page(
            viewport={"width": 1280, "height": 720}
        )

        await page.goto(f"file://{html_path}", wait_until="networkidle", timeout=30000)
        await asyncio.sleep(2)

        total_slides = 12
        pages_html = []

        for i in range(total_slides):
            # Переключаем на нужный слайд через JS
            await page.evaluate(f"""
                (function(){{
                  var slides = document.querySelectorAll('.slide');
                  slides.forEach(function(s){{ s.classList.remove('active','exit'); }});
                  slides[{i}].classList.add('active');
                }})();
            """)
            await asyncio.sleep(0.5)

            # Делаем скриншот каждого слайда
            screenshot = await page.screenshot(
                full_page=False,
                type="png",
                clip={"x":0,"y":0,"width":1280,"height":720}
            )
            pages_html.append(screenshot)
            print(f"  📸 Слайд {i+1}/{total_slides} захвачен")

        await browser.close()
        return pages_html, pdf_filename

# Запускаем захват слайдов
print("\n🎬 Захватываю слайды...")
slides_screenshots, pdf_filename = await make_pdf()

# =====================================================
# СБОРКА PDF ИЗ СКРИНШОТОВ
# =====================================================
print("\n📄 Собираю PDF из скриншотов...")
subprocess.run(["pip", "install", "-q", "Pillow", "fpdf2"], capture_output=True)

from PIL import Image
from fpdf import FPDF
import io

pdf = FPDF(orientation="L", unit="mm", format=(203.2, 114.3))  # 16:9 пропорция
pdf.set_auto_page_break(False)

for idx, screenshot_bytes in enumerate(slides_screenshots):
    img = Image.open(io.BytesIO(screenshot_bytes))

    # Сохраняем временный файл
    tmp_img = f"/tmp/slide_{idx:02d}.png"
    img.save(tmp_img, "PNG")

    pdf.add_page()
    pdf.image(tmp_img, x=0, y=0, w=203.2, h=114.3)
    print(f"  📄 Добавлен слайд {idx+1}/12")

pdf.output(pdf_filename)
print(f"\n✅ PDF создан: {pdf_filename}")
print(f"📦 Размер PDF: {os.path.getsize(pdf_filename):,} байт")
print(f"📊 Страниц: 12")

# =====================================================
# СКАЧИВАНИЕ ОБОИХ ФАЙЛОВ
# =====================================================
from google.colab import files

print("\n⬇️  Скачиваю HTML...")
files.download(html_filename)

print("⬇️  Скачиваю PDF...")
files.download(pdf_filename)

print("\n🎉 Готово! Скачаны файлы:")
print(f"   📄 {html_filename}")
print(f"   📋 {pdf_filename}")
```
